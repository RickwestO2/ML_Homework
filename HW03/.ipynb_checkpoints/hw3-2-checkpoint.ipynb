## (二) 針對信用卡交易資料，預測是否為詐騙的交易（class==1）

### 1. 讀入資料、切割資料（測試集佔 30%，訓練集佔 70%）

```python
import pandas as pd

df1= pd.read_csv("HW3_creditcard.csv")
df1
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
X=df1.drop(columns=['Class'])
Y=df1['Class']
x_train,x_test,y_train,y_test= train_test_split(X,Y,test_size=0.3,random_state=42)
```

### 2. 利用 Decision tree 進行預測，計算出 Accuracy, Recall, Precision,F1-Score 及 AUROC。

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, roc_auc_score

model = DecisionTreeClassifier(random_state=42)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)


Accuracy01=accuracy_score(y_test, y_pred)
Recall01=recall_score(y_test, y_pred)
Precision01=precision_score(y_test, y_pred)
F1Score01=f1_score(y_test, y_pred)
AUROC01=roc_auc_score(y_test, y_pred)

print("Accuracy: ",Accuracy01)
print("Recall: ",Recall01)
print("Precision: ",Precision01)
print("F1-Score: ",F1Score01)
print("AUROC: ",AUROC01)
```

### 3. 統計 class==0 及 class==1 的資料筆數，看是否類別間資料數量是否有很不平衡的現象。

```python
class0=df1[df1["Class"]==0].shape[0]
class1=df1[df1["Class"]==1].shape[0]
print("class==0: ",class0)
print("class==1: ",class1)
```

類別間資料不平衡

### 4. 為了要提高 recall 的數值，請：
### ⚫ 改變 Decision tree 模型中類別權重或訓練權重，計算新的結果，與之前結果比較。

```python
model = DecisionTreeClassifier(class_weight={0:1,1:10}, random_state=42)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

Accuracy02=accuracy_score(y_test, y_pred)
Recall02=recall_score(y_test, y_pred)
Precision02=precision_score(y_test, y_pred)
F1Score02=f1_score(y_test, y_pred)
AUROC02=roc_auc_score(y_test, y_pred)

print("Accuracy: ",Accuracy01)
print("Recall: ",Recall01)
print("Precision: ",Precision01)
print("F1-Score: ",F1Score01)
print("AUROC: ",AUROC01)

print("\n----After change weight----")
print("Accuracy: ",Accuracy02)
print("Recall: ",Recall02)
print("Precision: ",Precision02)
print("F1-Score: ",F1Score02)
print("AUROC: ",AUROC02)
```

### ⚫ 利用 imbalanced-learn 套件中 SMOTE 的方法來增量資料，計算新的結果，與之前結果比較。

```python
from imblearn.over_sampling import SMOTE
x_train_res, y_train_res = SMOTE(random_state=42).fit_resample(x_train, y_train)
model.fit(x_train_res, y_train_res)
y_pred = model.predict(x_test)

Accuracy03=accuracy_score(y_test, y_pred)
Recall03=recall_score(y_test, y_pred)
Precision03=precision_score(y_test, y_pred)
F1Score03=f1_score(y_test, y_pred)
AUROC03=roc_auc_score(y_test, y_pred)

print("Accuracy: ",Accuracy01)
print("Recall: ",Recall01)
print("Precision: ",Precision01)
print("F1-Score: ",F1Score01)
print("AUROC: ",AUROC01)

print("\n----After change weight----")
print("Accuracy: ",Accuracy02)
print("Recall: ",Recall02)
print("Precision: ",Precision02)
print("F1-Score: ",F1Score02)
print("AUROC: ",AUROC02)

print("\n----After SMOTE----")
print("Accuracy: ",Accuracy03)
print("Recall: ",Recall03)
print("Precision: ",Precision03)
print("F1-Score: ",F1Score03)
print("AUROC: ",AUROC03)
```

### 5. 改利用 xgboost 模型重新做 step 2。

```python
from xgboost import XGBClassifier

model = XGBClassifier(random_state=42)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

Accuracy04=accuracy_score(y_test, y_pred)
Recall04=recall_score(y_test, y_pred)
Precision04=precision_score(y_test, y_pred)
F1Score04=f1_score(y_test, y_pred)
AUROC04=roc_auc_score(y_test, y_pred)

print("Accuracy: ",Accuracy01)
print("Recall: ",Recall01)
print("Precision: ",Precision01)
print("F1-Score: ",F1Score01)
print("AUROC: ",AUROC01)

print("\n----After change weight----")
print("Accuracy: ",Accuracy02)
print("Recall: ",Recall02)
print("Precision: ",Precision02)
print("F1-Score: ",F1Score02)
print("AUROC: ",AUROC02)

print("\n----After SMOTE----")
print("Accuracy: ",Accuracy03)
print("Recall: ",Recall03)
print("Precision: ",Precision03)
print("F1-Score: ",F1Score03)
print("AUROC: ",AUROC03)

print("\n----After xgboost----")
print("Accuracy: ",Accuracy04)
print("Recall: ",Recall04)
print("Precision: ",Precision04)
print("F1-Score: ",F1Score04)
print("AUROC: ",AUROC04)
```
