## **1. 首先轉換成如下圖的 dataframe 的格式**

```python
import pandas as pd

df = pd.read_csv("HW1_data.csv")
df=df.pivot(index='0', columns='1', values='2')
column_order = ["Sex", "Age", "HR", "Height", "Weight", "BP"]
df = df[column_order]
# df
```

```python
new_df=df[column_order]
new_df.insert(0, "Name", df.index)
new_df.index=range(df.shape[0])
new_df.columns.name = None
new_df
```

## **2. 計算出 Age, HR, Height, Weight 及 BP 的平均值加於 dataframe 最下列。（請注意避免累加過程中出現 overflow 現象）**

```python
new_df =new_df.fillna(0)
new_df = new_df.astype({"Age":'int64', "HR":'int64', 'Height':'int64', 'Weight':'int64', 'BP':'int64'})
# new_df.info()
# new_df
```

```python
tmp={'Name':'Average','Sex':'','Age':new_df.mean().Age,'HR':new_df.mean().HR, 'Height':new_df.mean().Height, 'Weight':new_df.mean().Weight, 'BP':new_df.mean().BP}
df2=new_df.append(tmp, ignore_index=True)
df2
```

## **3. 利用程式找出並於螢幕列出各分項指標 feature（Age, HR, Height,Weight 及 BP）中之最大者的姓名（Name）。**

```python
for column in ['Age', 'HR', 'Height', 'Weight', 'BP']:
    max_value = new_df[column].max()
    max_name=new_df[new_df[column] == max_value]['Name'].values
    print("最大{}者是:{}, 數值為: {}".format(column,max_name,max_value))
```

## **4. 請繪製身高體重（Height, Weigth）之散佈圖（Scatter Plot），女生請以紅點標示，男生請以藍點標示。**

```python
import matplotlib.pyplot as plt
df3=df.dropna()
df3=df3.astype({"Age":'int64', "HR":'int64', 'Height':'int64', 'Weight':'int64', 'BP':'int64'})
plt.xlabel('Height')
plt.ylabel('Weight')
plt.scatter(df3[df3['Sex'] == 'F']['Height'], df3[df3['Sex'] == 'F']['Weight'],color='red')
plt.scatter(df3[df3['Sex'] == 'M']['Height'], df3[df3['Sex'] == 'M']['Weight'],color='blue')
plt.title('Height&Weight')
plt.show()
```

## **5. 繪製不同年齡（Age）區間的人數直方圖及圓餅圖（Pie Chart）。請依1-10, 11-20, 21-30…每 10 歲為一個區間統計。**

```python
import matplotlib.pyplot as plt
x=range(0,100,10)
x1=[]
for i in range(10):
    tmp=(str(x[i]+1)+"-"+str(x[i]+10))
    x1.append(tmp)

# print("x1:",x1)

list=[0] * 10
for j in range(new_df.shape[0]):
    if(new_df['Age'][j]%10>0):
        y=new_df['Age'][j]//10
        list[y]=list[y]+1
    else:
        y=new_df['Age'][j]//10 #10、20、30.....90
        list[y-1]=list[y-1]+1
y1=list    
# print("y1:",y1)

x2=[]
y2=[]
for k in range(0,len(y1)):
    if(y1[k]>0):
        y2.append(y1[k])
        x2.append(x1[k])

# print(x2)         
# print(y2)  

#Bar 
plt.bar(x2, y2)
plt.show()

#Pie chart
plt.pie(y2, radius=1.5, labels=x2,autopct='%.1f%%')
plt.show()
```
