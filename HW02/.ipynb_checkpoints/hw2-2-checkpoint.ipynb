# (二)使用 Linear Regression 預測在不同的時間，租借共享單車的人數預測 (count)

### 1. 讀入訓練資料 HW2_bike-sharing_train.csv。

```python
import pandas as pd

df1= pd.read_csv("HW2_bike-sharing_train.csv")
df1
```

```python
df2= pd.read_csv("HW2_bike-sharing_test.csv")
df2
```

### 2. 對日期欄位進行處理。

```python
def date_fun(data):
    tmp=pd.to_datetime(data["datetime"])
    #Xt["year"]=tmp.dt.year
    data.insert(0, "year", tmp.dt.year)
    data.insert(1, "month", tmp.dt.month)
    data.insert(2, "day", tmp.dt.day)
    data.insert(3, "hour", tmp.dt.hour)
    data.drop(columns=["datetime"], inplace=True)
    print(data.shape)
    return data
```

```python
df1=date_fun(df1)
df2=date_fun(df2)
```

```python
df1
```

```python
df2
```

```python
############################## train繪圖(df1)
```

```python
import matplotlib.pyplot as plt
columns = [col for col in df1.columns if col != 'count']
for col in columns:
    l = ['count']
    l.append(col) 
    bike_count = df1[l].groupby(by=col).sum().reset_index()
    bike_count.columns = [col,'count']
    display(bike_count )
    
    #plt.plot(bike_count[col], bike_count['count'], marker='o')
    plt.bar(bike_count[col], bike_count['count'])
    plt.xlabel(col)
    plt.ylabel('Total Count')
    plt.title('Total Count by '+col)
    plt.grid(True)
    plt.show()
```

```python
#####################################
```

```python
#分成 X,Y ,合併train.csv、test.csv做預處理
X=df1.drop(columns=['count'])
X=X.append(df2)
Y=df1['count']
print(X.shape,Y.shape)
```

```python
numerical_features = []
categorical_features = []

# splitting features into numeric and categoric.
'''
If feature has more than 35 categories we will consider it
as numerical_features, remaining features will be added to categorical_features.
'''
columns = [col for col in X.columns if col != 'count']
for col in columns:  
  if X[col].nunique() > 35:
    numerical_features.append(col) 
  else:
    categorical_features.append(col) 

print(f'Numerical Features : {numerical_features}')
print(f'Categorical Features : {categorical_features}')
```

```python
# we are going to replace the datapoints with upper and lower bound of all the outliers

def clip_outliers(X):
    for col in X[numerical_features]:
        # using IQR method to define range of upper and lower limit.
        q1 = X[col].quantile(0.25)
        q3 = X[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        
        # replacing the outliers with upper and lower bound
        X[col] = X[col].clip(lower_bound, upper_bound)
    return X
```

```python
# using the function to treat outliers
X = clip_outliers(X)
```

```python
#################################
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
#features_to_normalize = ['temp', 'atemp', 'humidity', 'windspeed']
features_to_normalize=[col for col in X.columns if col != 'count']
scaler = StandardScaler()
X[features_to_normalize]= scaler.fit_transform(X[features_to_normalize])
X
```

```python
print(X[:df1.shape[0]].shape)
print(X[:df2.shape[0]].shape)
```

```python
x=X[:df1.shape[0]]
y=Y
print(x.shape,y.shape)
```

### 3. 利用 regression 推測使用人數。產生準確率的指標。

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)
model = LinearRegression()
model.fit(x_train,y_train)

y_pred=model.predict(x_test)
y_pred[y_pred < 0] = 1 #負數設為1
plt.scatter(x_test.index, y_test, c='lightblue',alpha=0.5,label='y_test')
plt.scatter(x_test.index, y_pred, c='lightgreen',alpha=0.5,label='y_pred')

#plt.plot([min(x_test.index), max(x_test.index)], [min(y_test), max(y_test)], color='blue', linestyle='--',label='y_test')
#plt.plot([min(x_test.index), max(x_test.index)], [min(y_pred), max(y_pred)], color='green', linestyle='--',label='y_pred')
plt.legend()
plt.xlabel('index')
plt.ylabel('count')
plt.show()

# print("準確率的指標:")
# score =model.score(x_test,y_test)
# print('Score: ', score)
# print('Accuracy: ' + str(score*100) + '%')
```

```python
from sklearn.metrics import  r2_score
r2 = r2_score(y_test, y_pred)
print("R-squared (判定係數):", r2)
```

#### 準確率的指標:
#### R-squared, R2: R平方(R squared)又稱為判定係數(coefficient of determination)，是一種衡量回歸模型表現的指標

### 4. 請利用訓練後的模型預測測試資料 HW2_ bike-sharing_test.csv 的離職情況，並將結果存成 HW2_bike-sharing _test_sol.csv，儲存格式如下範例。該結果的準確率將佔此一題分數的 35%

```python
import pandas as pd

test=X[:df2.shape[0]] #HW2_ bike-sharing_test.csv
test
```

```python
new_pred=model.predict(test)
new_pred[new_pred < 0] = 1
new_pred=new_pred.astype(int)
new_df = pd.DataFrame({'count': new_pred})
new_df.to_csv("HW2_bike-sharing_test_sol.csv", index=False)
new_df
```

```python
plt.scatter(test.index, new_pred, c='lightblue',alpha=0.5,label='y_test')
plt.xlabel('index')
plt.ylabel('count')
plt.show()
```

```python
# (二)使用 Linear Regression 預測在不同的時間，租借共享單車的人數預測 (count)

### 1. 讀入訓練資料 HW2_bike-sharing_train.csv。

import pandas as pd

df1= pd.read_csv("HW2_bike-sharing_train.csv")
df1

df2= pd.read_csv("HW2_bike-sharing_test.csv")
df2

### 2. 對日期欄位進行處理。

def date_fun(data):
    tmp=pd.to_datetime(data["datetime"])
    #Xt["year"]=tmp.dt.year
    data.insert(0, "year", tmp.dt.year)
    data.insert(1, "month", tmp.dt.month)
    data.insert(2, "day", tmp.dt.day)
    data.insert(3, "hour", tmp.dt.hour)
    data.drop(columns=["datetime"], inplace=True)
    print(data.shape)
    return data

df1=date_fun(df1)
df2=date_fun(df2)

df1

df2

############################## train繪圖(df1)

import matplotlib.pyplot as plt
columns = [col for col in df1.columns if col != 'count']
for col in columns:
    l = ['count']
    l.append(col) 
    bike_count = df1[l].groupby(by=col).sum().reset_index()
    bike_count.columns = [col,'count']
    display(bike_count )
    
    #plt.plot(bike_count[col], bike_count['count'], marker='o')
    plt.bar(bike_count[col], bike_count['count'])
    plt.xlabel(col)
    plt.ylabel('Total Count')
    plt.title('Total Count by '+col)
    plt.grid(True)
    plt.show()

#####################################

#分成 X,Y ,合併train.csv、test.csv做預處理
X=df1.drop(columns=['count'])
X=X.append(df2)
Y=df1['count']
print(X.shape,Y.shape)

numerical_features = []
categorical_features = []

# splitting features into numeric and categoric.
'''
If feature has more than 35 categories we will consider it
as numerical_features, remaining features will be added to categorical_features.
'''
columns = [col for col in X.columns if col != 'count']
for col in columns:  
  if X[col].nunique() > 35:
    numerical_features.append(col) 
  else:
    categorical_features.append(col) 

print(f'Numerical Features : {numerical_features}')
print(f'Categorical Features : {categorical_features}')

# we are going to replace the datapoints with upper and lower bound of all the outliers

def clip_outliers(X):
    for col in X[numerical_features]:
        # using IQR method to define range of upper and lower limit.
        q1 = X[col].quantile(0.25)
        q3 = X[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        
        # replacing the outliers with upper and lower bound
        X[col] = X[col].clip(lower_bound, upper_bound)
    return X

# using the function to treat outliers
X = clip_outliers(X)

#################################

from sklearn.preprocessing import StandardScaler,MinMaxScaler
#features_to_normalize = ['temp', 'atemp', 'humidity', 'windspeed']
features_to_normalize=[col for col in X.columns if col != 'count']
scaler = StandardScaler()
X[features_to_normalize]= scaler.fit_transform(X[features_to_normalize])
X

print(X[:df1.shape[0]].shape)
print(X[:df2.shape[0]].shape)

x=X[:df1.shape[0]]
y=Y
print(x.shape,y.shape)

### 3. 利用 regression 推測使用人數。產生準確率的指標。

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)
model = LinearRegression()
model.fit(x_train,y_train)

y_pred=model.predict(x_test)
y_pred[y_pred < 0] = 1 #負數設為1
plt.scatter(x_test.index, y_test, c='lightblue',alpha=0.5,label='y_test')
plt.scatter(x_test.index, y_pred, c='lightgreen',alpha=0.5,label='y_pred')

#plt.plot([min(x_test.index), max(x_test.index)], [min(y_test), max(y_test)], color='blue', linestyle='--',label='y_test')
#plt.plot([min(x_test.index), max(x_test.index)], [min(y_pred), max(y_pred)], color='green', linestyle='--',label='y_pred')
plt.legend()
plt.xlabel('index')
plt.ylabel('count')
plt.show()

# print("準確率的指標:")
# score =model.score(x_test,y_test)
# print('Score: ', score)
# print('Accuracy: ' + str(score*100) + '%')


from sklearn.metrics import  r2_score
r2 = r2_score(y_test, y_pred)
print("R-squared (判定係數):", r2)

#### 準確率的指標:
#### R-squared, R2: R平方(R squared)又稱為判定係數(coefficient of determination)，是一種衡量回歸模型表現的指標

### 4. 請利用訓練後的模型預測測試資料 HW2_ bike-sharing_test.csv 的離職情況，並將結果存成 HW2_bike-sharing _test_sol.csv，儲存格式如下範例。該結果的準確率將佔此一題分數的 35%

import pandas as pd

test=X[:df2.shape[0]] #HW2_ bike-sharing_test.csv
test



new_pred=model.predict(test)
new_pred[new_pred < 0] = 1
new_pred=new_pred.astype(int)
new_df = pd.DataFrame({'count': new_pred})
new_df.to_csv("HW2_bike-sharing_test_sol.csv", index=False)
new_df

plt.scatter(test.index, new_pred, c='lightblue',alpha=0.5,label='y_test')
plt.xlabel('index')
plt.ylabel('count')
plt.show()



```
